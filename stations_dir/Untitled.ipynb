```python
def get_station_id(province, start_year, city):
    
    # Read the demand data for past 3 years
    df =  pd.read_csv("directory.csv")
    df1 = df[(df['HLY First Year'].notnull()) & (df['Province'] == str(province).upper()) & (df['HLY First Year'] <= start_year) & (df['HLY Last Year'] >= start_year )]
    df1 = df1[df1['Name'].str.contains(city.upper())]
    return df1
```
